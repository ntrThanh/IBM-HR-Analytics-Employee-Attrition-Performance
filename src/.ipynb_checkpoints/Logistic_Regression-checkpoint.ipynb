## Định nghĩa lại lớp phân loại Logistic Regression

```python
class LogisticRegression:
    def __init__(self, eta = 0.05, tol = 1e-4, maxcount = 1000):
        self.eta = eta
        self.tol = tol
        self.maxcount = maxcount
        self.X = None 
        self.y = None
        
    def fit(self, X, y):
        self.X = X
        self.y = y

    def sigmoid(self, s):
        return 1/(1 + np.exp(-s))

    def sigmoid_stable(self, s):
        if s >= 200:
            return 1
        if s < -200:
            return 0
    
        return 1/(1 + np.exp(-s))

    def caculate_w(self, X, y, w_init, eta, tol = 1e-4, max_count = 10000):
        w = [w_init]
        it = 0
        d = X.shape[1]
        N = X.shape[0]
        count = 0
        check_w_after = 20
        while count < max_count:
            # mix data for stochastic gradient descent method
            mix_id = np.random.permutation(N)
            for i in mix_id:
                xi = X[i, :].reshape(d, 1)
                yi = y[i]
                zi = self.sigmoid_stable(np.dot(w[-1].T, xi))
                w_new = w[-1] + eta*(yi - zi)*xi
                count += 1
                # stopping criteria
                if count%check_w_after == 0:
                    if np.linalg.norm(w_new - w[-check_w_after]) < tol:
                        return w
                w.append(w_new)
        return w
    
    def predict(self, X_test):
        self.w = self.caculate_w(self.X, self.y, self.w_init, self.eta, self.tol, self.maxcount)
        
        return []
```
