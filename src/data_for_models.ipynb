## Dữ liệu sau khi sử dụng One Hot Encoding và Label Encoding

Sử dụng thư viện `sklearn` để mã hóa one hot cho dữ liệu ban đầu

```python
X_encode = X.copy()
y_encode = y.copy()
```

```python
print(X_encode.shape)
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# define model encoding
one_hot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
label_encoder = LabelEncoder()
```

Encode với X

```python
columns_encode = ['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus', 'Over18', 'OverTime']
print("X_encode:")
print(X_encode)
print("-------------------")

# Dữ liệu sẽ được encode trong các cột bên trên
X_encoded_array = one_hot_encoder.fit_transform(X_encode[columns_encode])
print("X_encode[columns_encode]: ")
print(X_encode[columns_encode])
print("-------------------")

print("X_encoded_array: ")
print(X_encoded_array)
print("-------------------")

encoded_columns = one_hot_encoder.get_feature_names_out(columns_encode)
print("encoded_columns: ")
print(encoded_columns)
print("-------------------")

# Chuyển từ dạng nén, vector sang Dataframe cho dễ sử dụng hơn
X_encoded_df = pd.DataFrame(X_encoded_array, columns=encoded_columns, index=X_encode.index)
print("X_encoded_df:")
print(X_encoded_df)
print("-------------------")

# Xóa cột thừa có chữa kiểu string vì không sử dụng được
X_encode = pd.concat([X_encode.drop(columns=columns_encode, axis=1), X_encoded_df], axis=1)
print("X_encode: ")
print(X_encode)
print("-------------------")
```

```python
print(X_encode.shape)
```

```python
X_encode
```

Encode labels với y

```python
y_encode = label_encoder.fit_transform(y_encode)
```

```python
y_encode
```

```python
print(f'Hình dạng của X sau khi mã hóa: {X_encode.shape}')
print(f'Kiểu của X sau khi mã hóa: {type(X_encode)}')
```

Dữ liệu đã được encode:

## Dữ liệu sau khi giảm chiều

### Giảm chiều bằng PCA mà không chuẩn hóa

#### Sẽ thực hiện giảm chiều dữ liệu 80, 90, 95% covariance được giữ lại (giảm trước khi chia train/test)

Kí hiệu pca_(covariance) tức là kí hiệu với việ thực hiện giảm chiều với việc giữ lại phần trăm covarince

```python
X_pca = X_encode.copy()
y_pca = y_encode.copy()
```

```python
from sklearn.decomposition import PCA

pca_80_percent = PCA(n_components=0.8)
pca_90_percent = PCA(n_components=0.9)
pca_95_percent = PCA(n_components=0.95)

X_pca_80 = pca_80_percent.fit_transform(X_pca)
X_pca_90 = pca_90_percent.fit_transform(X_pca)
X_pca_95 = pca_95_percent.fit_transform(X_pca)
```

Chia dữ liệu, mặc định chia với tỉ lệ Train / Test = 7 / 3

```python
from sklearn.model_selection import train_test_split
X_train_after_pca_80, X_test_after_pca_80, y_train_after_pca_80, y_test_after_pca_80 = train_test_split(X_pca_80, y_pca, test_size=0.3, random_state=42)

X_train_after_pca_90, X_test_after_pca_90, y_train_after_pca_90, y_test_after_pca_90 = train_test_split(X_pca_90, y_pca, test_size=0.3, random_state=42)

X_train_after_pca_95, X_test_after_pca_95, y_train_after_pca_95, y_test_after_pca_95 = train_test_split(X_pca_95, y_pca, test_size=0.3, random_state=42)
```

Chia trước khi giảm chiều

```python
X_pca = X_encode.copy()
y_pca = y_encode.copy()

X_train_before_pca, X_test_before_pca, y_train_before_pca, y_test_before_pca = train_test_split(X_pca, y_pca, test_size=0.3, random_state=42)
```

```python
from sklearn.decomposition import PCA

pca_20_percent = PCA(n_components=0.8)
pca_50_percent = PCA(n_components=0.90)
pca_80_percent = PCA(n_components=0.95)

X_train_before_pca_80 = pca_80_percent.fit_transform(X_train_before_pca)
X_test_before_pca_80 = pca_80_percent.transform(X_test_before_pca)

X_train_before_pca_90 = pca_50_percent.fit_transform(X_train_before_pca)
X_test_before_pca_90 = pca_50_percent.transform(X_test_before_pca)

X_train_before_pca_95 = pca_20_percent.fit_transform(X_train_before_pca)
X_test_before_pca_95 = pca_20_percent.transform(X_test_before_pca)
```

***Thực hiện với lần lượt các train, test trên***

#### Thực hiện giảm chiều giữ lại 2, 4, 8, 10

Ở phần này chỉ thực hiện mỗi việc chia rồi mới giảm chiều

```python
X_pca = X_encode.copy()
y_pca = y_encode.copy()
```

```python
from sklearn.model_selection import train_test_split
X_train_pca_n2, X_test_pca_n2, y_train_pca_n2, y_test_pca_n2 = train_test_split(X_pca, y_pca, test_size=0.3, random_state=42)

X_train_pca_n4, X_test_pca_n4, y_train_pca_n4, y_test_pca_n4 = train_test_split(X_pca, y_pca, test_size=0.3, random_state=42)

X_train_pca_n8, X_test_pca_n8, y_train_pca_n8, y_test_pca_n8 = train_test_split(X_pca, y_pca, test_size=0.3, random_state=42)

X_train_pca_n10, X_test_pca_n10, y_train_pca_n10, y_test_pca_n10 = train_test_split(X_pca, y_pca, test_size=0.3, random_state=42)
```

Ở đây kí hiệu pca_n2 tức là thực hiện pca với n = 2

```python
from sklearn.decomposition import PCA

pca_n2 = PCA(n_components=2)
pca_n4 = PCA(n_components=4)
pca_n8 = PCA(n_components=8)
pca_n10 = PCA(n_components=10)

X_train_pca_n2 = pca_n2.fit_transform(X_train_pca_n2)
X_test_pca_n2 = pca_n2.transform(X_test_pca_n2)

X_train_pca_n4 = pca_n4.fit_transform(X_train_pca_n4)
X_test_pca_n4 = pca_n4.transform(X_test_pca_n4)

X_train_pca_n8 = pca_n8.fit_transform(X_train_pca_n8)
X_test_pca_n8 = pca_n8.transform(X_test_pca_n8)

X_train_pca_n10 = pca_n10.fit_transform(X_train_pca_n10)
X_test_pca_n10 = pca_n10.transform(X_test_pca_n10)
```

# Mô tả dữ liệu

Về dữ liệu nguyên thủy: `X_origin`, `y_origin`

Về dữ liệu đã được mã hóa: `X_encode`, `y_encode`

Về dữ liệu đã được chuẩn hóa: `X_standard`, `y_standard`

Đối với dữ liệu đã được giảm chiều mà không chuẩn hóa:
- Thực hiện giữ lại `n%` covariance giảm sau đó chia train/test: `X_train_after_pca_n`, `X_test_after_pca_n`, `y_train_after_pca_n`, `y_test_after_pca_n`. với n ở đây là 80, 90, 95%

- Thực hiện giữ lại `n%` covariance giảm trước đó chia train/test: `X_train_before_pca_n`, `X_test_before_pca_n`, `y_train_before_pca_n`, `y_test_before_pca_n`. với n ở đây là 80, 90, 95%

- Thực hiện giữ lại `n` thành phần: `X_train_pca_n<n>`, `X_test_pca_n<n>`, `y_train_pca_n<n>`, `X_test_pca_n<n>` với n trong dấu `<n>` là số phần tử được giữ lại với n ở đây là 2, 4, 8, 10.

Giảm chiều bằng PCA với dữ liệu đã được chuẩn hóa trước đó:
- Giữ lại `n%` covariance: `X_train_before_pca_<n>_std`, `X_test_before_pca_<n>_std`, `y_train_before_pca_<n>_std`, `y_test_before_pca_<n>_std`  


- Giữ lại `n` với `n` được đặt trong `<_>` thành phần: `X_train_pca_n<n>_std`, `X_test_pca_n<n>_std`, `y_train_pca_n<n>_std`, `y_test_pca_n<n>_std` với n thuộc 2, 4, 8, 10

Giảm chiều với LDA đã được chuẩn hóa: `X_train_lda_std`, `X_test_lda_std`, `y_train_lda_std`, `y_test_lda_std`

Giảm chiều vói LDA chưa chuẩn hóa: `X_train_lda`, `X_test_lda`, `y_train_lda`, `y_test_lda`
